## Gene Set Activations

```python
def get_gene_sets(filename):
    gene_sets = []
    with open(filename, "r") as f:
        for line in f:
            tokens = line.split()
            if float(tokens[2]) >= 0.2:
                gene_sets.append((tokens[0], float(tokens[1]), float(tokens[2])))
    gene_sets.sort(key=lambda x: (-x[2], -x[1]))
    return gene_sets

def print_gene_sets(gene_sets):
    print("Top 10 %\tAvg. Activation\tGene Set")
    for gene_set, act, top_10 in gene_sets:
        print("{}\t{}\t{}".format(top_10, act, gene_set))
```

### Combo (trained independently)

```python
combo_gene_sets = get_gene_sets("combo_weights.txt")
print_gene_sets(combo_gene_sets)
```

### Lapatinib (trained independently)

```python
lap_gene_sets = get_gene_sets("lap_weights.txt")
print_gene_sets(lap_gene_sets)
```

### Trastuzumab (trained independently)

```python
trast_gene_sets = get_gene_sets("trast_weights.txt")
print_gene_sets(trast_gene_sets)
```

### Cell Line Lapatinib (trained independently)

```python
cl_gene_sets = get_gene_sets("cell_lines_weights.txt")
print_gene_sets(cl_gene_sets)
```

### Gene Set Intersections

```python
combo = set(map(lambda x:x[0], combo_gene_sets))
lap = set(map(lambda x:x[0], lap_gene_sets))
trast = set(map(lambda x:x[0], trast_gene_sets))
cell_line = set(map(lambda x:x[0], cl_gene_sets))
```

```python
print("Intersection of Combo and Lapatinib:")
print(combo.intersection(lap))

print("\nIntersection of Combo and Trastuzumab:")
print(combo.intersection(trast))

print("\nIntersection of all:")
print(combo.intersection(lap, trast))

print("\n Intersection of Lapatinib (clinical) and Lapatinib (cell line)")
print(lap.intersection(cell_line))
```

### Multi-Task No Regularizer

```python
gene_sets = get_gene_sets("noreg_NN_activations.txt")
print_gene_sets(gene_sets)
```

### Multi-Task With L1 Regularization (0.0012)

```python
l1_gene_sets = get_gene_sets("NN_activations.txt")
print_gene_sets(l1_gene_sets)
```

```python
gs_1 = set(map(lambda x:x[0], gene_sets))
gs_2 = set(map(lambda x:x[0], l1_gene_sets))
print("\nIntersection of No Reg. and Reg.:")
intersection = gs_1.intersection(gs_2)
print(intersection)
print(len(intersection))
```

```python
with open("top_sets.txt", "w") as f:
    for gene_set, _, _ in l1_gene_sets:
        f.write(gene_set + "\n")
        
with open("noreg_top_sets.txt", "w") as f:
    for gene_set, _, _ in gene_sets:
        f.write(gene_set + "\n")
```
